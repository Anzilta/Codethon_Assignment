```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
```

```python
Data=pd.read_csv("CarPrice_Assignment.csv")
Data.head()
```

```python
Data.shape
```

```python
Data.info()
```

```python
Data.isnull().sum()
```

```python
Data.duplicated().sum()
```

```python
 le=LabelEncoder()
Str_col=Data.select_dtypes(include=["string"])
for col in Str_col:
    Data[col]=le.fit_transform(Data[col])
Data.info()
```

```python
Data.drop("car_ID",axis=1,inplace=True)
```

```python
Sc=StandardScaler()
x=Data.drop("price",axis=1)
y=Data["price"]

x_scaled=Sc.fit_transform(x)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

# Linear Regressoin

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr_pred=lr.predict(X_test)

print("Linear Regression")
print("R2 Score:", r2_score(Y_test, lr_pred))
print("MAE:", mean_absolute_error(Y_test, lr_pred))
print("MSE:", mean_squared_error(Y_test, lr_pred))
```

# Decision Tree 

```python
dt=DecisionTreeRegressor(random_state=42)

dt.fit(X_train,Y_train)
```

```python
dt_pred=dt.predict(X_test)
```

```python
print("Decision Tree")
print("R2 Score:", r2_score(Y_test, dt_pred))
print("MAE:", mean_absolute_error(Y_test, dt_pred))
print("MSE:", mean_squared_error(Y_test, dt_pred))
```

 # Random Forest

```python
rf = RandomForestRegressor(random_state=42)

rf.fit(X_train, Y_train)
```

```python
rf_pred = rf.predict(X_test)

print("Random Forest")
print("R2 Score:", r2_score(Y_test, rf_pred))
print("MAE:", mean_absolute_error(Y_test, rf_pred))
print("MSE:", mean_squared_error(Y_test, rf_pred))
```

# Gradient Boosting

```python
gr=GradientBoostingRegressor(random_state=42)
gr.fit(X_train,Y_train)
```

```python
gr_pred=gr.predict(X_test)

print("Gradient Boosting")
print("R2 Score=",r2_score(Y_test,gr_pred))
print("MAE:", mean_absolute_error(Y_test, gr_pred))
print("MSE:", mean_squared_error(Y_test, gr_pred))
```

# Support Vector

```python
svr=SVR()

svr.fit(X_train,Y_train)
```

```python
svr_pred=svr.predict(X_test)

print("Support Vencctor Machine")
print("R2 Score=",r2_score(Y_test,svr_pred))
print("MAE:", mean_absolute_error(Y_test, svr_pred))
print("MSE:", mean_squared_error(Y_test, svr_pred))
```

# Model Comparisons


```python
Models={"LinearRegression":LinearRegression(),
        "DecisionTree":DecisionTreeRegressor(random_state=42),
        "RandomForest":RandomForestRegressor(random_state=42),
        "GradientBoosting":GradientBoostingRegressor(random_state=42),
        "SVR":SVR()
       }
Best_Model=None
Best_r2score=0.0
Best_mse=10000
Best_mae=10000

for name,model in Models.items():
    model.fit(X_train,Y_train)
    y_pred=model.predict(X_test)
    print(name)
    R2_Score=r2_score(Y_test,y_pred)
    print("R2_score:",round(R2_Score,4) )              
    MAE=mean_absolute_error(Y_test, y_pred)
    print("MAE:",round(MAE,4))                        
    MSE=mean_squared_error(Y_test, y_pred)
    print("MSE:",round(MSE,4))  
    print("\n")
    if R2_Score > Best_r2score:
               Best_Model = name
               Best_r2score = R2_Score
               Best_mae = MAE
               Best_mse = MSE
print("Best_Model=",Best_Model)
print("r2score=",Best_r2score)
print("MAE=",Best_mae) 
print("MSE=",Best_mse)
       


    
        
```

<H2>Best Model</H2>
<h4>The Random Forest Regressor achieved the highest R² Score and the lowest MSE and MAE among all the models. This indicates that it predicts car prices more accurately than the other algorithms. Random Forest performs well because it combines multiple decision trees, reducing overfitting and effectively capturing complex relationships between the features and the target variable.</h4>

# Hyper Tunning

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
rf = RandomForestRegressor(random_state=42)

grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

grid.fit(X_train, Y_train)
```

```python
print("Best Parameters:")
print(grid.best_params_)
```

```python
best_rf = grid.best_estimator_

y_pred = best_rf.predict(X_test)

print("R2 Score :", round(r2_score(Y_test, y_pred),4))
print("MAE :", round(mean_absolute_error(Y_test, y_pred),4))
print("MSE :", round(mean_squared_error(Y_test, y_pred),4))
```
